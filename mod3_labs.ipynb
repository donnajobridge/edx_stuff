```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.set_xlabel('area')
ax.set_ylabel('perimeter')
ax.set_zlabel('asymmetry')
ax.scatter(df.area, df.perimeter, df.asymmetry, c='r')
```
