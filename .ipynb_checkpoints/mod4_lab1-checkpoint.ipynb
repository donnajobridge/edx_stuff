```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

from mpl_toolkits.mplot3d import Axes3D
from plyfile import PlyData, PlyElement
from sklearn.decomposition import PCA
```

```python
plt.style.use('ggplot')
```

```python
reduce_factor = 100
```

```python
plyfile=PlyData.read('../DAT210x/Module4/Datasets/stanford_armadillo.ply')

armadillo=pd.DataFrame({
    'x':plyfile['vertex']['z'][::reduce_factor],
    'y':plyfile['vertex']['x'][::reduce_factor],
    'z':plyfile['vertex']['y'][::reduce_factor]
})

```

```python
def do_PCA(df, svd_solver):
    pca=PCA(n_components=2, svd_solver=svd_solver)
    pca.fit(df)
    T=pca.transform(df)  
    return T
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.set_title('armadillo 3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter(armadillo.x, armadillo.y, armadillo.z, c='g', marker='.', alpha=.75)
```

```python
%timeit pca=do_PCA(armadillo,'full')
```

```python
pca=do_PCA(armadillo,'full')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_title('Full PCA')
ax.scatter(pca[:,0], pca[:,1],c='blue', marker='.', alpha=.75)
```

```python
%timeit rpca=do_PCA(armadillo, 'randomized')
```

```python
rpca=do_PCA(armadillo, 'randomized')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_title('Randomized PCA')
ax.scatter(rpca[:,0], rpca[:,1], c='red', marker='.', alpha=.75)
```
